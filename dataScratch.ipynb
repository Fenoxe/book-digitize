```python
import numpy as np
import pandas as pd
from pickle import load
```

```python
with open('firstSpreadData.p','rb') as f:
    df = load(f)
```

```python
df['conf'] = df['conf'].astype(int)
```

```python
display(df.head())
df.info()
```

```python
print(df['text'].value_counts().to_string())
```

```python
valConfDf = df[df['conf'] > 0]
```

```python
display(valConfDf.head())
valConfDf.info()
```

```python
valTextDf = valConfDf[(valConfDf['text'] != '') & (valConfDf['text'] != ' ')]
```

```python
display(valTextDf.head())
valTextDf.info()
```

```python
df2 = valTextDf
```

```python
df2.sort_values('width',ascending=False)
```

```python
df2.sort_values('conf',ascending=True)
```

```python
print(df2.sort_values('conf')[['left','top','width','height','conf','text']].to_string())
```
